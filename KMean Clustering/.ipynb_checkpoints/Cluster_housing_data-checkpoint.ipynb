Data cleaning is required for this dataset, example houses with no bedrooms and no bathrooms. Zero value in yr_renovated and sqft_basement also distort their mean and medium value. During analysis also found that one house with 33 bedrooms can be an one off outlier. 

```python
#The outlier is numerically defined as Lower outlier = Q1-k(Q3-Q1) Upper outlier = Q3+k(Q3-Q1)
#Q1, Q2, Q3, are 1st quartile, median (2nd quartile), and 3rd quartile, respectively. k = 1.5.
Quant = data.quantile([0.25,0.5,0.75])
Quant
```

```python
#only apply outlier cleaning on sqft, price, bedroom feature
sqft_list = data.filter(regex='sqft|price|bedrooms').columns.to_list()
data_filter = data.copy()

for sqft in sqft_list:
    
    Q1_rate = Quant[sqft][0.25]
    Q3_rate = Quant[sqft][0.75]
    k=1.5
    IQR_rate = Q3_rate-Q1_rate
    Lower_out_rate = Q1_rate-k*(IQR_rate)
    Upper_out_rate = Q3_rate+k*(IQR_rate)
    print(sqft)
    print('Lower Outlier = ', Lower_out_rate, 'Upper Outlier = ', Upper_out_rate)
    data_filter = data_filter[(data_filter[sqft]>Lower_out_rate) & (data_filter[sqft]<Upper_out_rate)].copy()

print('before filter:', data.shape)
print('after filter:',data_filter.shape)
```

```python
print("Summary Statistics:\n")
sumary = data_filter.describe(percentiles=[])
print(np.transpose(np.round(sumary)))
```

```python
fig, axes = plt.subplots(5, 4, figsize=(20, 20))
for ax, feature in zip(axes.flat, data.columns):
    sns.distplot(data_filter[feature], ax=ax, kde=False)
```

```python
data_filter.filter(regex='sqft')
#from the data, we found sqft_living is = sqft_above + sqft_basement
```

```python
#perform feature engineering to sum total sqft of the house as square footage of the home + lot
data_filter['total_sqft'] = data_filter['sqft_living'] + data_filter['sqft_lot']
data_filter['total_sqft15'] = data_filter['sqft_living15'] + data_filter['sqft_lot15']

#created total rooms columns to replace bedroom and bathroom
data_filter['total_rooms'] = data_filter['bedrooms'] + data_filter['bathrooms']

#create total_score column to replace condition and grade
data_filter['total_score'] = data_filter['condition'] * data_filter['grade']

#create house age as of year 2020. House age = 2020 - yr_built. If renovated, house age = 2020 - yr_renovated
data_filter['house_age'] = np.where(data_filter['yr_renovated']==0, (2020 - data_filter['yr_built']), 
                                    (2020 - data_filter['yr_renovated']))

data_featured = data_filter.drop(columns=['sqft_lot', 'sqft_living','sqft_lot15','sqft_living15',
                                          'sqft_basement','sqft_above','bedrooms','bathrooms','condition',
                                          'grade', 'yr_built', 'yr_renovated', 'waterfront'])
```

```python
data_featured.head()
```

```python
#split data into 50/50 for validation
X_train, X_test = train_test_split(data_featured, test_size=0.5, random_state=42)
```

```python
# Standarize features
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
clusters = 10
sse = {}

for i in range(2, clusters):
    
    kmeans = KMeans(n_clusters=i,init="k-means++", n_init = 10, max_iter= 200, random_state=42, n_jobs=-1)
    kmeans.fit(X_train_std) 

    goodness = silhouette_score(X_train_std,kmeans.predict(X_train_std)).round(3)
    print("No Clusters =", i," Silhouette = ",goodness)
    
    sse[i] = kmeans.inertia_ #SSE of each point to its centroid
    
plt.figure(figsize=(10,6))
plt.plot(list(sse.keys()), list(sse.values()))
plt.title("Number of clusters vs SSE")
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
#choose number of custer based on elbow method with 6 number of cluster
kmeans = KMeans(n_clusters=6,init="k-means++", n_init = 10, max_iter= 100, random_state=42, n_jobs=-1)
kmeans.fit(X_train_std) 

#annotate cluster label on test dataset 
col_list = X_test.columns.to_list()
test_df = pd.DataFrame(X_test_std, columns=col_list)
test_df["Cluster"] = kmeans.predict(X_test_std)
# find number of members in each cluster
count = test_df.groupby('Cluster')['Cluster'].count()
# find cluster averages
cluster_means = np.round(test_df.groupby(['Cluster']).mean(),decimals=1)
```

```python
#test dataset with cluster annotation
test_df.head()
```

```python
cluster_means['Count'] = count
pd.DataFrame(np.transpose(cluster_means))
```

```python
#put table into dataframe, reshape for barplot
cluster_df = pd.DataFrame(np.transpose(cluster_means)).iloc[:8,:] #drop count
cluster_df = cluster_df.reset_index()
cluster_df = cluster_df.rename(columns={'index': 'Features'})
clus_df_melt = pd.melt(cluster_df, id_vars="Features", var_name="Cluster", value_name="Values")

plt.figure(figsize=(10,6))
ax = sns.barplot(data=clus_df_melt,x="Cluster",y="Values",hue="Features")
```

Defining the 6 clusters: <br/>
Cluster 0 - High number of total rooms and floor, low house age <br/>
Cluster 1 - High house age and high score <br/>
Cluster 2 - Low price, low number of flow, low number of total rooms and low total score <br/>
Cluster 3 - High number of floors, low total_sqft, low total_sqft15 and low house age <br/>
Cluster 4 - High total_sqft, high total_sqft15, high total score <br/>
Cluster 5 - High view and high price
